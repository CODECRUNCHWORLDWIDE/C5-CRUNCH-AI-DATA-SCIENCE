```python
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
import kagglehub
import pandas as pd
```

```python
# Download latest version
path = kagglehub.dataset_download("akashdeepkuila/bakery")
```

```python
!ls {path}
```

# load data

```python
data = pd.read_csv(path+"/Bakery.csv")
data
```

# preprocess data

```python
transaction_arr = []
for d in data.iloc():
  if d["TransactionNo"] > len(transaction_arr):
    transaction_arr.append([])
  transaction_arr[-1].append(d["Items"])
```

## encode transactions

```python
te = TransactionEncoder()
```

```python
te_arr = te.fit(transaction_arr).transform(transaction_arr)
```

```python
df = pd.DataFrame(te_arr, columns=te.columns_)
df
```

# modelling

```python
freq_items = apriori(df, min_support=0.03, use_colnames=True)
```

# analysis

```python
association_rules(freq_items, metric="confidence", min_threshold=0.2)
```

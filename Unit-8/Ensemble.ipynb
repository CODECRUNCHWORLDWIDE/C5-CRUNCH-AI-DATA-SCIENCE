```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv("/content/drive/MyDrive/B-CODECRUNCH/spotify.csv")
data
```

```python
X = data[["energy","valence"]]
y = data["playlist_genre"]
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,test_size=0.1)
```

# Bagging

```python
base_estimator = DecisionTreeClassifier()
```

```python
model = BaggingClassifier(
    estimator=base_estimator,
    n_estimators=100,
    max_samples=1.0,
    max_features=1.0,
)
```

```python
model.fit(Xtrain,ytrain)
```

```python
ypred = model.predict(Xtest)
```

```python
accuracy_score(ytest,ypred)
```

# Boosting

```python
model = AdaBoostClassifier()
```

```python
model.fit(Xtrain,ytrain)
```

```python
ypred = model.predict(Xtest)
```

```python
accuracy_score(ytest,ypred)
```

# Stacking

```python
base_estimators = [
    ('rf', RandomForestClassifier(n_estimators=50, random_state=42)),
    ('svm', SVC(kernel='linear', probability=True, random_state=42)),
    ('dt', DecisionTreeClassifier(random_state=42))
]
```

```python
meta_model = LogisticRegression()
```

```python
stacking_model = StackingClassifier(estimators=base_estimators, final_estimator=meta_model, cv=5)
```

```python
stacking_model.fit(Xtrain, ytrain)
```

# Exploratory Data Analysis (EDA)
Exploratory Data Analysis (EDA) is a critical step in data analysis to understand the dataset's structure, relationships, and anomalies.


## Import Necessary Libraries
Ensure libraries like pandas, numpy, or other required libraries are imported:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

## Data Collection
 - collect the data:
  - in our case we find our data from the cloud

- load the data:

```python
data = load_iris(as_frame=True)
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

## Understanding the Data

- display the table:

```python
X # print(X)
```

- display a few instances:

```python
X.head(1)
```

- display size:

```python
X.shape
```

- display data types:

```python
X.info(verbose=True)
```

- get the mean of a specific column:

```python
X['petal length (cm)'].mean()
```

- get the median of a specifc column:

```python
X["sepal length (cm)"].median()
```

- get the range of a specifc column

```python
X["petal width (cm)"].max() - X["petal width (cm)"].min()
```

- concatenate the two subsets into the original bigger dataset

```python
data = pd.concat([X,y],axis=1)
data
```

## Analyze The Data

- scatter plot
  - good if you want to view the relationship between two numerical values

```python
sns.scatterplot(data,x="sepal length (cm)",y="sepal width (cm)",hue="target")
```

- histogram plot
  - good to see the spread of categorical data
    - good to see the balance a categorical feature

```python
sns.histplot(data,x="target")
```

- box plot
  - good for view numerical data across different features

```python
sns.boxplot(data,x="target", y="petal length (cm)")
```

- FacetGrid,

```python
g = sns.FacetGrid(data, col='target')
g.map(sns.histplot, 'sepal width (cm)')
```

- pair plot

```python
sns.pairplot(data,hue="target")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.svm import SVR

from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
data = pd.read_csv("/content/drive/MyDrive/B-CODECRUNCH/spotify.csv")
```

```python
data #print(data)
```

```python
X = data[["liveness","tempo","instrumentalness","acousticness"]]
y = data['valence']
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,train_size=0.9)
```

```python
Xtrain
```

```python
reg = SVR()
reg.fit(Xtrain, ytrain)
ypred = reg.predict(Xtest)
```

```python
pd.DataFrame(ypred)
```

```python
mean_absolute_error(ytest, ypred)
```

```python
mean_squared_error(ytest, ypred)
```

```python
R = Ridge(0.001)
R.fit(Xtrain, ytrain)
ypred2 = R.predict(Xtest)
```

```python
pd.DataFrame(ypred)
```

```python
mean_absolute_error(ytest, ypred2)
```

```python
mean_squared_error(ytest, ypred2)
```

```python
L = Lasso(0.001)
L.fit(Xtrain, ytrain)
ypred3 = L.predict(Xtest)
```

```python
mean_absolute_error(ytest, ypred3)
```

```python
mean_squared_error(ytest, ypred3)
```

```python
EN = ElasticNet(0.001)
EN.fit(Xtrain, ytrain)
ypred4 = EN.predict(Xtest)
```

```python
mean_absolute_error(ytest, ypred4)
```

```python
mean_squared_error(ytest, ypred4)
```
